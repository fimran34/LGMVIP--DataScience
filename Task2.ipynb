```python
import cv2
import matplotlib.pyplot as plt
```

```python
origimage = cv2.imread("v.jpg")
```

```python
grayimage = cv2.cvtColor(origimage, cv2.COLOR_BGR2GRAY) 
```

```python
invertimage = cv2.bitwise_not(grayimage)
```

```python
blurimg=cv2.GaussianBlur(invertimage, (71,71),0)
```

```python
invertblurimg=cv2.bitwise_not(blurimg)
```

```python
sketchimage=cv2.divide(grayimage,invertblurimg, scale=256.0)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,6,1)
plt.imshow(cv2.cvtColor(origimage, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,6,2)
plt.imshow(grayimage,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,6,3)
plt.imshow(invertimage,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,6,4)
plt.imshow(blurimg,cmap="gray")
plt.axis("off")
plt.title("Blur Image")
plt.subplot(1,6,5)
plt.imshow(invertblurimg,cmap="gray")
plt.axis("off")
plt.title("Blur Invert Image")
plt.subplot(1,6,6)
plt.imshow(sketchimage,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
